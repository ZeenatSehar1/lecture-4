Ask the user for a password in a while loop. Allow 3 tries, and if the correct password is
entered, print "Access granted"
, otherwise print "Blocked".


```python
password=12345
i=1
while i<=3:
  if int(input("Enter password: "))==password:
    print("Access granted")
    break
  i+=1

else:
    print("Blocked")
```

From a list of words, create a dictionary where key is the word and value is the length, but
only for words longer than 4 letters.


```python
words = ["apple", "cat", "banana", "dog", "elephant", "sun"]

result = {}

for word in words:
    if len(word) > 4:
        result[word] = len(word)

print(result)
```

Given a string, build a dictionary using comprehension where the keys are characters and
values are their frequency in the string

```python
text = "programming"

freq = {ch: text.count(ch) for ch in text}

print(freq)
```

Create a list from 100 to 150 using list comprehension where prime numbers stay as-is, and
non-primes become "Not Prime".

```python
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return n > 1

result = [n if is_prime(n) else "Not Prime" for n in range(100, 151)]

print(result)
```

**List Categorizer** Given a list of integers:
l = list(range(1,21))
Separate them into 3 lists:
Even
Odd
Multiples of 5
Print all 3 lists at the end

```python
l = list(range(1, 21))

even = []
odd = []
multiples_of_5 = []

for num in l:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

    if num % 5 == 0:
        multiples_of_5.append(num)

print("Even numbers:", even)
print("Odd numbers:", odd)
print("Multiples of 5:", multiples_of_5)
```

**FizzBuzz Dictionatry** Make a dictionary for numbers 1 to 30 with:
"Fizz" if divisible by 3
"Buzz" if divisible by 5
"FizzBuzz" if divisible by both
"None" otherwise


```python
fizzbuzz_dict = {}

for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        fizzbuzz_dict[i] = "FizzBuzz"
    elif i % 3 == 0:
        fizzbuzz_dict[i] = "Fizz"
    elif i % 5 == 0:
        fizzbuzz_dict[i] = "Buzz"
    else:
        fizzbuzz_dict[i] = "None"

for key, value in fizzbuzz_dict.items():
    print(f"{key}: {value}")
```
